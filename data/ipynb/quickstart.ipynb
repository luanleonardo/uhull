```python
import matplotlib.pyplot as plt

from uhull.alpha_shape import get_alpha_shape_polygons

points = [
    (0.0, 0.0),
    (0.0, 1.0),
    (1.0, 1.0),
    (1.0, 0.0),
    (0.5, 0.25),
    (0.5, 0.75),
    (0.25, 0.5),
    (0.75, 0.5),
]
polygons = get_alpha_shape_polygons(coordinates_points=points)
x, y = zip(*points)
plt.plot(x, y, ".")
plt.savefig("../img/set_of_points.png", bbox_inches="tight")
```

```python
polygons = get_alpha_shape_polygons(coordinates_points=points)

x, y = zip(*points)
plt.plot(x, y, ".")
for polygon in polygons:
    x, y = zip(*polygon)
    plt.plot(x, y)

plt.savefig("../img/concave_hull_points_set.png", bbox_inches="tight")
```

```python
polygons = get_alpha_shape_polygons(coordinates_points=points, alpha=2 * 1.5)

x, y = zip(*points)
plt.plot(x, y, ".")
for polygon in polygons:
    x, y = zip(*polygon)
    plt.plot(x, y)

plt.savefig(
    "../img/concave_hull_doubling_default_alpha_value.png",
    bbox_inches="tight",
)
```

```python
def manhattan_distance(coord1, coord2):
    return abs(coord1[0] - coord2[0]) + abs(coord1[1] - coord2[1])


polygons = get_alpha_shape_polygons(
    coordinates_points=points, distance=manhattan_distance
)

x, y = zip(*points)
plt.plot(x, y, ".")
for polygon in polygons:
    x, y = zip(*polygon)
    plt.plot(x, y)

plt.savefig(
    "../img/concave_hull_with_manhattan_distance.png", bbox_inches="tight"
)
```
